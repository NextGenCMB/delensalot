# Interactive minimal working example - check
This notebook demonstrates how to access the initialized filters, iterators, etc of a jobs before running a reconstruction. This is helpful if you want check that everything initialized correctly before you start your run.

```python
from IPython.display import clear_output
from datetime import datetime
import inspect
import hashlib
import psutil

import healpy as hp
import numpy as np
import matplotlib.pyplot as plt
import os
from os.path import join as opj

import delensalot
from delensalot.run import run
from delensalot.utils import camb_clfile
from delensalot.utility.utils_hp import Alm, almxfl, alm_copy, gauss_beam

from delensalot.config.visitor import transform, transform3d
from delensalot.core.opfilt.opfilt_handler import QE_transformer, MAP_transformer
from delensalot.config.transformer.lerepi2dlensalot import l2delensalotjob_Transformer, l2T_Transformer
from delensalot.config.metamodel.dlensalot_mm import DLENSALOT_Model, DLENSALOT_Analysis
from delensalot.core.iterator.iteration_handler import iterator_transformer

from delensalot.config.metamodel.dlensalot_mm import DLENSALOT_Model, DLENSALOT_Qerec, DLENSALOT_Itrec, DLENSALOT_Computing, DLENSALOT_Noisemodel, DLENSALOT_Analysis, DLENSALOT_Mapdelensing, DLENSALOT_Simulation
```

```python
configuration_model = DLENSALOT_Model()
print(configuration_model)
```

# Generate sim

```python
job_id = 'generate_sim'
job_model = transform3d(configuration_model, job_id, l2delensalotjob_Transformer())

clear_output(wait=True)
print("Cell finished {}".format(datetime.now().strftime("%H:%M:%S")))

job_model.__dict__
```

# MAP lensrec

```python
configuration_model = DLENSALOT_Model()
job_id = 'MAP_lensrec'
job_model = transform3d(configuration_model, job_id, l2delensalotjob_Transformer())

clear_output(wait=True)
print("Cell finished {}".format(datetime.now().strftime("%H:%M:%S")))

job_model.__dict__
```

```python
MAP_filters = transform(job_model, MAP_transformer())
# MAP_filters.
```

```python
MAP_filter = transform(job_model, MAP_filters())
MAP_filter.__dict__
```

```python
iterator_model = transform(job_model, iterator_transformer(job_model, 0, job_model.dlensalot_model))
```
